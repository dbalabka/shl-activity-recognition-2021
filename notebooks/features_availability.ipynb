```python
import pandas as pd
from shl.prepare import normalize_epoch_time
pd.set_option('display.max_columns', None)
```

```python
train_label = normalize_epoch_time(pd.read_parquet('../data/train/Label.parquet'), 'epoch_time')
test_label = normalize_epoch_time(pd.read_parquet('../data/test/Label_idx.parquet'), 'epoch_time')
validate_label = normalize_epoch_time(pd.read_parquet('../data/validate/Label.parquet'), 'epoch_time')
```

```python
train_location = normalize_epoch_time(pd.read_parquet('../data/train/Location.parquet'), 'epoch_time')
train_cells = normalize_epoch_time(pd.read_parquet('../data/train/Cells.parquet'), 'epoch_time')
train_gps = normalize_epoch_time(pd.read_parquet('../data/train/GPS.parquet'), 'Epoch time [ms]')
train_wifi = normalize_epoch_time(pd.read_parquet('../data/train/WiFi.parquet'), 'Epoch time [ms]')

validate_location = normalize_epoch_time(pd.read_parquet('../data/validate/Location.parquet'), 'epoch_time')
validate_cells = normalize_epoch_time(pd.read_parquet('../data/validate/Cells.parquet'), 'epoch_time')
validate_gps = normalize_epoch_time(pd.read_parquet('../data/validate/GPS.parquet'), 'Epoch time [ms]')
validate_wifi = normalize_epoch_time(pd.read_parquet('../data/validate/WiFi.parquet'), 'Epoch time [ms]')

test_location = normalize_epoch_time(pd.read_parquet('../data/test/Location.parquet'), 'epoch_time')
test_cells = normalize_epoch_time(pd.read_parquet('../data/test/Cells.parquet'), 'epoch_time')
test_gps = normalize_epoch_time(pd.read_parquet('../data/test/GPS.parquet'), 'Epoch time [ms]')
test_wifi = normalize_epoch_time(pd.read_parquet('../data/test/WiFi.parquet'), 'Epoch time [ms]')
```

```python
train_location['location available'] = 1
train_cells['cells available'] = 1
train_gps['gps available'] = 1
train_wifi['wifi available'] = 1

validate_location['location available'] = 1
validate_cells['cells available'] = 1
validate_gps['gps available'] = 1
validate_wifi['wifi available'] = 1

test_location['location available'] = 1
test_cells['cells available'] = 1
test_gps['gps available'] = 1
test_wifi['wifi available'] = 1
```

```python
train_availability = train_label.merge(train_location[['epoch_time_id', 'location available']].drop_duplicates(), how='left')\
                                .merge(train_cells[['epoch_time_id', 'cells available']].drop_duplicates(), how='left')\
                                .merge(train_gps[['epoch_time_id', 'gps available']].drop_duplicates(), how='left')\
                                .merge(train_wifi[['epoch_time_id', 'wifi available']].drop_duplicates(), how='left')

validate_availability = validate_label.merge(validate_location[['epoch_time_id', 'location available']].drop_duplicates(), how='left')\
                                      .merge(validate_cells[['epoch_time_id', 'cells available']].drop_duplicates(), how='left')\
                                      .merge(validate_gps[['epoch_time_id', 'gps available']].drop_duplicates(), how='left')\
                                      .merge(validate_wifi[['epoch_time_id', 'wifi available']].drop_duplicates(), how='left')

test_availability = test_label.merge(test_location[['epoch_time_id', 'location available']].drop_duplicates(), how='left')\
                              .merge(test_cells[['epoch_time_id', 'cells available']].drop_duplicates(), how='left')\
                              .merge(test_gps[['epoch_time_id', 'gps available']].drop_duplicates(), how='left')\
                              .merge(test_wifi[['epoch_time_id', 'wifi available']].drop_duplicates(), how='left')
```
